```python
import numpy as np
import pandas as pd
import seaborn as sns
import pickle
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_data = pd.read_csv('heart_disease_data.csv')
```

```python
#number of rows and columns in the dataset
heart_data.shape
```

```python
#getting statistical measures about the data
heart_data.describe()
```

```python
#checking distribution of target variable(how many 1's and 0's)
heart_data['target'].value_counts()
```

```python
#age and strength of heart, target
sns.displot(x='age', hue = 'target', data = heart_data)
```

```python
X = heart_data.drop('target',axis = 1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(X.head())
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify = Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
# training the logistic regression model with training data
model.fit(X_train, Y_train)
```

```python
# accuracy score on training data
X_train_predict = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predict, Y_train)
```

```python
print('Accuracy on training data: ', training_data_accuracy)
```

```python
# accuracy score on test data
X_test_predict = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predict, Y_test)
```

```python
print('Accuracy on training data: ', test_data_accuracy)
```

```python
input_data = (65,1,0,135,254,0,0,127,0,2.8,1,1,3)

# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)

print(prediction)

if (prediction[0]==0):
  print('The Person does not have a Heart Disease')
else:
  print('The Person has Heart Disease')
```

```python
# Save the model
with open('heart_disease_prediction_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```
